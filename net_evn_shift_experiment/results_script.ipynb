```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import os
import numpy as np

import seaborn as sns

os.getcwd()

categories = pd.read_csv('Documents/CENT/simulations_results_categories.csv')
categories_env = pd.read_csv('Documents/CENT/simulations_env_shift_results_categories.csv')
categories_hub = pd.read_csv('Documents/CENT/simulations_hub_sim_results.csv')
env_shift_sim = pd.read_csv('Documents/CENT/env_shift_sim_results.csv')


data = pd.read_csv('Downloads/simulations_results.csv')  #get direction of simulation files!

data.head(20000/50+2)
results = list(data.columns)[2:]


colors = ['lightblue','darkblue','r','orange', 'brown', 'g', 'darkgreen', 'y','k','orange', 'pink', 'brown', 'magenta']
networks = ["max_avg_bet", "max_avg_clust", "max_max_bet", "max_max_clos",
            "max_var_cons", "min_avg_bet", "min_avg_clust", "min_max_clos", "line", "hub", "clique"]
```

```python
# hub + clique CSA CLA
hub = ['hub_hearer', 'hub_speaker']

plt.figure(1, figsize=(15,10))

for i, network in enumerate(hub):
    plt.plot(categories_hub[categories_hub.iloc[:,0] == network]["CSA_mean"][0:-200], colors[2*i], label=network+"_CSA")
    plt.plot(categories_hub[categories_hub.iloc[:,0] == network]["CLA_mean"], colors[2*i], label=network+"_CLA")
  
plt.plot(categories[categories.iloc[:,0] == "hub"]["CLA_mean"], colors[4], label="Hub_CLA")
plt.plot(categories[categories.iloc[:,0] == "clique"]["CLA_mean"], colors[6], label="Clique_CLA")
    
plt.legend(loc='upper left', frameon=False)
```

```python
#environment shift

plt.figure(1, figsize=(15,10))
plt.ylim(0, 1)

for i, network in enumerate(networks):
    plt.plot(env_shift_sim[env_shift_sim.iloc[:,0] == network]["CLA_mean"], colors[i], label=network)
    plt.plot(env_shift_sim[env_shift_sim.iloc[:,0] == network]["CSA_mean"][0:-100], colors[i])
    
plt.legend(loc='upper left', frameon=False)
```
